# Predicting Carbon Dioxide Emission of a New Car Model based on its Engine Size using Simple Linear Regression

### Importing Required Libraries

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Downloading the Data

```python
df = pd.read_csv(r'C:\Users\user\Desktop\Data Science\FuelConsumptionCo2.csv')
df.head()
```

### Data Exploration

```python
df.describe()
```

### Extracting Necessary Columns

```python
cdf = df[['MODEL', 'ENGINESIZE', 'CO2EMISSIONS']]
cdf = cdf.set_index('MODEL')
cdf.head()
```

### Plotting Engine Size and CO2 Emissions

```python
cdf.hist()
```

### Plotting CO2 Emissions vs Engine Size

```python
plt.figure()
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS)
plt.xlabel('ENGINE SIZE')
plt.ylabel('CO2 EMISSIONS')
plt.title('Relationship between CO2 Emission & Engine Size')
```

### Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
X = cdf[['ENGINESIZE']]
y = cdf[['CO2EMISSIONS']]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state=0)
```

### Simple Linear Regression Model

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
print("Coefficient = ", reg.coef_)
print("Intercept = ", reg.intercept_)
```

### Plotting the Fit Line over the Data

```python
plt.figure()
plt.scatter(X_train.ENGINESIZE, y_train.CO2EMISSIONS)
plt.xlabel('ENGINE SIZE')
plt.ylabel('CO2 EMISSIONS')
plt.title('Relationship between CO2 Emission & Engine Size')
plt.plot(X_train.ENGINESIZE, reg.coef_[0][0]*X_train.ENGINESIZE + reg.intercept_[0], '-r')
```

### Model Accuracy

```python
reg.score(X_test, y_test)
```

### Predicting CO2 Emission for a given Engine Size

```python
reg.predict([[7.5]])
```

### Evaluation Metrics

```python
from sklearn.metrics import r2_score
y_hat_test = reg.predict(X_test)
print("MAE = ", np.mean(np.absolute(y_test - y_hat_test)))
print("RMSE = ", np.sqrt(np.mean((y_test - y_hat_test)**2)))
print("R2 Score = ", r2_score(y_test, y_hat_test))
```
